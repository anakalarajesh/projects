subplot

```python
students = ["Rajesh", "Pavan", "Sunil", "Ravi"]
marks = [90, 85, 80, 75]
attendance = [95, 88, 92, 85]

# Figure with 1 row and 2 columns

plt.subplot(1, 2, 1)     # 1 row and 2 columns 1 plot
plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.subplot(1, 2, 2)    # 1 row and 2 columns 1 plot
plt.plot(students, attendance, marker='o')
plt.title("Attendance")
plt.xlabel("Students")
plt.ylabel("Percentage")

plt.suptitle("student information")
plt.tight_layout() #(automatically adjust spacing b/w two subplots)
plt.show()
```

```python
students = ["Rajesh", "Pavan", "Sunil", "Ravi"]
marks = [90, 85, 80, 75]
attendance = [95, 88, 92, 85]

# Figure with 1 row and 2 columns

plt.subplot(2, 1, 1)     # 1 row and 2 columns 1 plot
plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.subplot(2, 1, 2)    # 1 row and 2 columns 1 plot
plt.plot(students, attendance, marker='o')
plt.title("Attendance")
plt.xlabel("Students")
plt.ylabel("Percentage")

plt.suptitle("student information")
plt.tight_layout() #(automatically adjust spacing b/w two subplots)
plt.show()
```

```python
hours=[3,4,5,6,7,8,9]
marks=[30,40,50,60,70,80,90]
colors=["red","green","blue","yellow","pink","black","orange"]
plt.scatter(hours,marks,c=colors)
plt.title("Study Hours vs Marks",color="red",fontsize=20)
plt.xlabel("Study Hours",color="blue")
plt.ylabel("Marks",color="blue")
plt.show()
```

```python
hours=[3,4,5,6,7,8,9,10,11]
marks=[30,35,45,52,60,72,85,92,98]

color=[20,30,40,50,60,70,80,90,100]
plt.scatter(hours,marks,c=color,cmap='viridis')
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.colorbar()
plt.show()
```

```python
hours=[3,4,5,6,7,8,9,10,11]
marks=[30,35,45,52,60,72,85,92,98]

size=[20,30,60,150,260,370,480,590,700]
plt.scatter(hours,marks,s=size)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
hours=[3,4,5,6,7,8,9,10,11]
marks=[30,35,45,52,60,72,85,92,98]

size=[20,30,60,150,260,370,480,590,700]
plt.scatter(hours,marks,s=size,alpha=0.5)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

Histogram (bins means data divided in a histogram)

```python
hours=[3,4,5,6,7,8,9]
marks=[30,40,50,60,70,80,90]
plt.hist(marks, bins=5)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

pie chart

```python
subject=["pyhton","Sql","java","webtechnology"]
marks=[90,80,70,50]
plt.pie(marks,labels=subject)
plt.legend()
plt.show()
```

```python
subject=["pyhton","Sql","java","webtechnology"]
marks=[90,80,70,50]
mycolors=["red","yellow","pink","magenta"]
myexplode=[0.2,0,0.1,0]
plt.pie(marks,labels=subject,colors=mycolors,explode=myexplode)
plt.show()
```

```python
subject=["pyhton","Sql","java","webtechnology"]
marks=[90,80,70,50]
myexplode=[0.2,0,0.1,0]
plt.pie(marks,labels=subject,explode=myexplode,shadow=True,autopct="%1.1f%%")
plt.show()  #autopct-->automatically displays values on pie slices, usually as percentages,
```
