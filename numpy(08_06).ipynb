np.take() is used to extract elements from an array using specified indices

```python
arr=np.array([10,20,30,40,50])
result=np.take(arr,[0,2,4])
print(result)
```

```python
arr = np.array([10, 20, 30, 40])  # repeated indices
result = np.take(arr, [1, 1, 3])
print(result)
```

```python
arr = np.array([[1, 2],    # 2D array
                [3, 4]])
result = np.take(arr, [0, 3])
print(result)
```

```python
arr = np.array([[10, 20, 30],  # using axis, axis=1 means columns, takes column 0 and column 2
                [40, 50, 60]])
result = np.take(arr, [0, 2], axis=1)
print(result)
```

swapaxes --> swap two axes   np.swapaxes(array,axis1,axis2)
array___>input array,
axis1____>first axis,
axis2___>secound axis

```python
arr=np.array([[1,2,3],
              [4,5,6]])
result=np.swapaxes(arr,0,1)
print(result)
```
